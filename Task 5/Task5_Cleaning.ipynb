```python
import pandas as pd

df = pd.read_csv("gender_submission.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
before = df.shape[0]
df = df.drop_duplicates()
after = df.shape[0]

print("Rows before:", before)
print("Rows after:", after)
```

```python
# Convert Survived to integer
df['Survived'] = df['Survived'].astype(int)

# Convert PassengerId to integer
df['PassengerId'] = df['PassengerId'].astype(int)
```

```python

# Survival label
df['Survival_Status'] = df['Survived'].map({1: 'Survived', 0: 'Not Survived'})
```

1. Loaded the Titanic dataset and inspected its structure using head() and info().
2. Identified missing values using isnull().sum().
3. Filled missing Age values using median to avoid skew.
4. Filled missing Embarked values using mode.
5. Dropped Cabin column due to excessive missing data.
6. Removed duplicate rows and verified row count.
7. Converted Survived and PassengerId to integer types.
8. Created Survived status feature for better analysis.
9. Saved the cleaned dataset for further modeling.
