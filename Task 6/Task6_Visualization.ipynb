```python
import pandas as pd

# Load dataset
df = pd.read_csv("world_happiness.csv")   # use your file name

# View first few rows
df.head()

# Check column names and datatypes
df.info()
```

```python
import matplotlib.pyplot as plt

# Sort and select top 10
top10 = df.sort_values(by="Score", ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(top10["Country or region"], top10["Score"])
plt.xlabel("Country")
plt.ylabel("Happiness Score")
plt.title("Top 10 Happiest Countries")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.savefig("top10_happiest_countries.png", bbox_inches='tight')
plt.show()
```

```python
# Line Chart  Happiness Score Trend by Rank
df_sorted = df.sort_values("Overall rank")

plt.figure(figsize=(10,6))
plt.plot(df_sorted["Overall rank"], df_sorted["Score"])
plt.xlabel("Overall Rank")
plt.ylabel("Happiness Score")
plt.title("Happiness Score Trend by Rank")
plt.grid(True)
plt.savefig("line_happiness_trend.png", bbox_inches='tight')
plt.show()
```

```python
# Histogram  Distribution of Happiness Scores
plt.figure(figsize=(8,5))
plt.hist(df["Score"], bins=15)
plt.xlabel("Happiness Score")
plt.ylabel("Frequency")
plt.title("Distribution of Happiness Scores")
plt.grid(True)
plt.savefig("happiness_distribution.png", bbox_inches='tight')
plt.show()
```

```python
# Scatter Plot  GDP vs Happiness Score
plt.figure(figsize=(8,6))
plt.scatter(df["GDP per capita"], df["Score"])
plt.xlabel("GDP per capita")
plt.ylabel("Happiness Score")
plt.title("GDP per Capita vs Happiness Score")
plt.grid(True)
plt.savefig("scatter_gdp_vs_happiness.png", bbox_inches='tight')
plt.show()
```

```python
# Key Insight
top_country = top10.iloc[0]

plt.figure(figsize=(10,6))
plt.bar(top10["Country or region"], top10["Score"])
plt.scatter(top_country["Country or region"], top_country["Score"])
plt.text(0, top_country["Score"], "Highest Score", color="black")
plt.title("Top 10 Happiest Countries (Highlight)")
plt.xticks(rotation=45)
plt.show()
```

Insights from Visualizations

1. Finland ranks as the happiest country, significantly outperforming others in happiness score.
2. Happiness score shows a strong positive correlation with GDP per capita.
3. The distribution of happiness scores is slightly right-skewed, with fewer countries achieving very high scores.
